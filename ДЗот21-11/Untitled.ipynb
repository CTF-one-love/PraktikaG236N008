```python
import numpy as np
import matplotlib.pyplot as plt
```

### Код для примера нейронной сети, про девочку из первого видео.

```python
def act(x):
    return 0 if x < 0.5 else 1

def go(house, rock, attr):
    x = np.array([house, rock, attr])
    w11 = [0.3, 0.3, 0]
    w12 = [0.4, -0.5, 1]
    weight1 = np.array([w11, w12])  # матрица 2x3
    weight2 = np.array([-1, 1])     # вектор 1х2

    sum_hidden = np.dot(weight1, x)       # вычисляем сумму на входах нейронов скрытого слоя
    print("Значения сумм на нейронах скрытого слоя: "+str(sum_hidden))

    out_hidden = np.array([act(x) for x in sum_hidden])
    print("Значения на выходах нейронов скрытого слоя: "+str(out_hidden))

    sum_end = np.dot(weight2, out_hidden)
    y = act(sum_end)
    print("Выходное значение НС: "+str(y))

    return y

house = 1
rock = 0
attr = 1

res = go(house, rock, attr)
if res == 1:
    print("Ты мне нравишься")
else:
    print("Созвонимся")
```

### Код для первого примера из видео 2. 

```python
N = 5  # количество образов

# моделируем образы для класса 1
x1 = np.random.random(N) # случайные величины по оси х1
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] # моделируется как x1 + случайное отклонение
C1 = [x1, x2] # формируем двумерный список, состоящий из набора этих (случайных) точек

# моделируем образы для класса 1
x1 = np.random.random(N) 
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 # минусы - чтобы наша точка x2 была ниже прямой
C2 = [x1, x2] 

# формирование разделяющей прямой 
f = [0, 1] 

# задаём веса для наших 
w = np.array([-0.5, 0.5])
# перебор всех образов для класса C2
for i in range(N):
    x = np.array([C2[0][i], C2[1][i]])
    y = np.dot(w, x) # вычисляем выходное значение
    if y >= 0:
        print("Класс C1")
    else:
        print("Класс C2")
# отображаем точки для этих классов
plt.scatter(C1[0][:], C1[1][:], s=10, c='green')
plt.scatter(C2[0][:], C2[1][:], s=10, c='orange')
# отображаем прямую
plt.plot(f)
plt.grid(True)
plt.show()
```

### Код из примера 2, второго видео.

```python
N = 5 # количество образов
b = 3 # сдвиг  по оси x2

# формируем значения x1, x2
x1 = np.random.random(N)
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] + b # с добвалением b
C1 = [x1, x2]

x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 + b
C2 = [x1, x2]

# формирование разделяющей прямой c + b
f = [0+b, 1+b]
# Определяем коэффициенты омега 2 и 3
w2 = 0.5
w3 = -b*w2
# обозначаем все весовые коэффициенты
w = np.array([-w2, w2, w3])
# Пропускаем через нашу сеть все образы класса c1
for i in range(N):
    x = np.array([C1[0][i], C1[1][i], 1])
    y = np.dot(w, x)
    if y >= 0:
        print("Класс C1")
    else:
        print("Класс C2")

# отображаем точки для этих классов
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f)
plt.grid(True)
plt.show()
```

### Код из примера 3, для второго видео.

```python
# активационная функция
def act(x):
    return 0 if x <= 0 else 1
# основная функция запуска нейронной сети
def go(C):
    x = np.array([C[0], C[1], 1])
    w1 = [1, 1, -1.5] # веса для первого нейрона
    w2 = [1, 1, -0.5] # веса для второго нейрона
    w_hidden = np.array([w1, w2]) # объединяем веса в матрицу
    w_out = np.array([-1, 1, -0.5]) # оределение весов для выходного нейрона

    sum = np.dot(w_hidden, x) # cумма весов на каждом нейроне
    out = [act(x) for x in sum] # пропускаем церез активацонную функцию
    out.append(1) # добавляем bios
    out = np.array(out)

    sum = np.dot(w_out, out)
    y = act(sum) # выходное значение нейронной сети
    return y

C1 = [(1,0), (0,1)]
C2 = [(0,0), (1,1)]
# вывод полученных значений
print( go(C1[0]), go(C1[1]) )
print( go(C2[0]), go(C2[1]) )
```
